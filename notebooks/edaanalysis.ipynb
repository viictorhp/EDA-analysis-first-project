## 2. Limpieza y transformación de los datos

Voy a eliminar las columnas "height" y "weight", ya que están duplicadas pero con otras unidades de medida. Estas mismas columnas están en kilogramos (weight_kg) y centímetros (height_cm) y me voy a quedar con ellas mejor para este análisis.

```python
crossfit_df.drop(["height", "weight"], axis = 1, inplace=True)
```

Voy a renombrar algunas columnas para que se entiendan mas facilmente y voy a ponerlas en un formato de nomenclatura genérico.

```python
# Función de regex para añadir una "_" en medio de dos palabras y poner todo en minúscula
import re

crossfit_df.columns = [re.sub(r'([a-z])([A-Z])', r'\1_\2', col).lower() for col in crossfit_df.columns]
```

```python
crossfit_df = crossfit_df.rename(columns={
    "country_of_origin_name": "country",
    "region_name": "region",
    "overall_rank" : "rank",
    "overall_score" : "score"
})
crossfit_df.columns
```

#### TRATAMIENTO DE VALORES NULOS 

Las variables "affiliate_id" y "affiliate_name" son las que cuentan con mayor porcentaje de valores nulos en mi dataset. 

Estas variables hacen referencia al nombre del sitio donde entrenan estos atletas, lo cuál considero innecesario, ya que, cuento con el país y la región para establecer relaciones y hacer estudios demográficos y como variable de identificación voy a usar los nombres y la variable "competitor_id".

Por tanto voy a eliminar también estas dos variables.

```python
crossfit_df.drop(["affiliate_id", "affiliate_name"], axis = 1, inplace=True)
```

Centrándonos en la variable "country" voy a sustituir los valores nulos por "NoCountry" para indicar que no disponemos de este valor. Considero que esto es más beneficioso para nuestro posterior análisis demográfico, ya que, tendremos en cuenta también a estos atletas.

```python
crossfit_df["country"] = crossfit_df["country"].fillna("NoCountry")
```

```python
crossfit_df[crossfit_df["country"] == "NoCountry"]
```

#### TRATAMIENTO DE VALORES ATÍPICOS

Como hemos visto antes para la variable "rank" existían algunos valores atípicos, es decir, negativos para algunos atletas.
Voy a transformar estos valores negativos en valores nulos para que así sigan formando parte del análisis y sean tratados como valores nulos.

```python
crossfit_df.loc[crossfit_df["rank"] < 0, "rank"] = np.nan
```

```python
rank_nan = crossfit_df[crossfit_df["rank"].isna()]
rank_nan[["competitor_id", "competitor_name", "rank"]]
```

Tenemos 85 valores nulos (80 valores que ya eran nulos + 5 valores que eran negativos y acabo de transformalos en nulos)

Voy a transformar el tipo de las variables de edad (age) y "competitor_id" para que sean tratados como valores integer.

```python
crossfit_df["age"] = crossfit_df["age"].astype(int)
crossfit_df["competitor_id"] = crossfit_df["competitor_id"].astype(int)
crossfit_df.dtypes
```

Compruebo que la variable "competitor_id" sólo cuenta con valores numéricos para la identificación de los atletas y que las variables "competitor name" , "country" y "region" sólo cuentan con valores de tipo object.

```python
# Intento convertir la columna a numérica y busco valores no convertibles (NaN)
no_numericos = crossfit_df[pd.to_numeric(crossfit_df["competitor_id"], errors="coerce").isna()]

# Muestro los casos con valores no numéricos (si existen)
print(no_numericos[["competitor_id"]])
```

```python
columns = ["competitor_name", "country", "region"]

# Selecciono filas donde el valor no es de tipo string y muestro los casos (si existen)
for col in columns:
    no_texto = crossfit_df[~crossfit_df[col].apply(lambda x: isinstance(x, str))]
    print(f"Valores no texto en la columna {col}:")
    print(no_texto[[col]])
```

Sí que se cumplen las condiciones mencionadas anteriormente, por tanto, no es necesario tratar ninguna fila en específico.
